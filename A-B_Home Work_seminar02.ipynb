## Задание 1 
Сделайте приоритезацию гипотез из предыдущего урока с помощью ICE

## ozon.ru

Ключевые метрики:

1. Доход от новых пользователей
2. Количество новых регистраций пользователей
3. Частота повторных покупок новых пользователей
4. Стоимость предоставляемой скидки
5. Оценка удовлетворенности клиентов новых пользователей

Гипотеза № 1 

Если предоставить персональные рекомендации по продуктам новым пользователям на основе их истории просмотров и покупательского поведения, сможем повысить показатель дохода от новых пользователей на n%.
Группа новых пользователей получает персонализированные рекомендации.
Тестируем на новых пользователях, совершивших хотя бы одну покупку на сайте

Метрика: доход от новых пользователей


Гипотеза № 2 

Если предложить 10% скидку за первую покупку для новых пользователей, можем повысить количество регистраций новых пользователей на 15%.
Для обеих груп даём код на 10% скидку за первую покупку.
Тестируем на новых пользователях, которые раньше не делали покупок на сайте.

Метрика: количество новых регистраций пользователей.

Гипотеза № 3

Если внедрить предложение программы лояльности, которая будет поощрять покупки баллами, можем увеличить частоту повторных покупок новыми пользователями на n%.
В каждой группе коздаем программу лояльности для новых пользователей, чтобы получать вознаграждение за каждую покупку баллами.
Тестируем на новых пользователях, совершивших хотя бы одну покупку на сайте.

Метрика: частота повторных покупок новыми пользователями

Гипотеза № 4

Если ввести предложение персональных скидок на основе поведения пользователей, снизим стоимость предоставляемой скидки.
В каждой группе анализируем поведение пользователей и предоставляем персональные скидки на основе их предпочтений, истории покупок.
Тестируем на всех пользователях, совершивших хотя бы одну покупку на сайте.

Метрика: Стоимость предоставленной скидки

Гипотеза № 5

Если создать обслуживание клиентов в связи с активностью Солнца, сможем улучшить оценку удовлетворенности новых пользователей в пвериоды повышенной солнечной активности.
В каждой из групп активно связываемся с новыми пользователями, чтобы решить любые проблемы, которые у них могут возникнуть всвязи с повыщенной метеочувствительностью. Так же связываемся с представителями гелиобиологии. 
Тестируем на овых пользователях, совершивших хотя бы одну покупку на сайте.

Метрика: оценка удовлетворенности клиентов новых пользователей

## Задание 2.

Составьте шаблон дизайна эксперимента для гипотезы, которая набрала больше всего баллов в практическом задании предыдущего урока

Так как в ICE -тестировании наибольшее количество набрала гипотеза № 2 на её основании и делаем шаблон дизайна эксперимента

Задача: Увеличить количество регистраций новых пользователей на ozon.ru

Ключевы метрики:

1. Доход от новых пользователей
2. Количество новых регистраций пользователей
3. Частота повторных покупок новых пользователей
4. Стоимость предоставляемой скидки
5. Оценка удовлетворенности клиентов новых пользователей

Метод:

Проведя A/B-тестирование, случайным образом разделив новых пользователей на две группы: одна группа получит код скидки 10 % на свою первую покупку, а другая группа не получит никакой скидки. Измерем перечисленные выше показатели для обеих групп и сравним результаты через месяц

Ожидаемый результат:

Ожидается значительное увеличение регистраций новых пользователей, получивших код скидки, что приведет к увеличению дохода при минимальном влиянии на стоимость предлагаемой скидки
